```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="South Korea")



geolocoder = Nominatim(user_agent = 'South Korea')
# geocoding 함수 정의
def geocoding_final(address):
    # geolocator = Nominatim(user_agent="South_Korea")
    location = geolocoder.geocode(address, timeout=None)
    while location is None:
        location = geolocoder.geocode(address, timeout=None)
    return location.latitude, location.longitude

# 결과를 저장할 빈 리스트 생성
pj_exfinal = []

# 모든 주소에 대해 geocoding2 함수 실행
for i in df['geocoding_juso4']:
    crd = geocoding_final(i)
    # 리스트로 변환
    place = list(crd)
    # 리스트에 추가
    pj_exfinal.append(place)

pj_exfinal
```

```python
import geopandas as gpd
from shapely.geometry import Point
# Sample coordinates
coordinates = pj_exfinal    

# Create a GeoDataFrame with the coordinates
geometry = [Point(xy) for xy in coordinates]
gdf = gpd.GeoDataFrame(geometry=geometry)

# Set the coordinate reference system (CRS) for the GeoDataFrame
gdf.crs = "EPSG:4326"  # WGS84 latitude-longitude projection
# EPSG:3857
# Save the GeoDataFrame as a shapefile
gdf.to_file("/Users/sinhuijin/mini_project/address_spots.shp")
```

```python
from folium import Map, Marker
# 그 다음, 지도를 생성합니다. Map() 함수에는 중심 좌표와 확대/축소 수준 등을 입력할 수 있습니다. 
# 이번 예시에서는 서울을 중심으로 지도를 그리고, 축적을 11로 지정하겠습니다.
seoul_map = Map(location=[37.566535, 126.9779692], zoom_start=11)
# 이제, Marker() 함수를 이용하여 좌표를 지도 위에 표시합니다. 
# Marker() 함수에는 좌표와 팝업 텍스트 등을 입력할 수 있습니다.
# 이번 예시에서는 pj_exfinal 리스트의 모든 좌표값을 지도 위에 표시하고, 팝업 텍스트로는 해당 좌표의 인덱스를 사용하겠습니다.

for idx, coord in enumerate(pj_exfinal):
    Marker(location=coord, popup=idx).add_to(seoul_map)

#     마지막으로, 지도를 html 파일로 저장합니다.
seoul_map.save('/Users/sinhuijin/mini_project/seoul_map.html')
```
